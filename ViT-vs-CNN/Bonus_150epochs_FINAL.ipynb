```python
from google.colab import drive
drive.mount('/content/drive')

import json
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

sns.set_style('whitegrid')
plt.rcParams['figure.figsize'] = (12, 6)
plt.rcParams['font.size'] = 10

BASE_PATH = '/content/drive/Shareddrives/Sieci_Neuronowe/Projekt_1'

with open(f'{BASE_PATH}/long_training_vit_best_150ep.json', 'r', encoding='utf-8') as f:
    data = json.load(f)

cifar  = data['cifar10']
inet   = data['imagenet']
summ   = data['summary']

print('Dane załadowane!')
print(f"Model: {summ['model']}, {summ['epochs']} epok, 32×32")
print(f"CIFAR-10    best: {summ['best_cifar_acc']:.2f}%  @ epoch {summ['best_cifar_epoch']}")
print(f"Imagenette  best: {summ['best_inet_acc']:.2f}%  @ epoch {summ['best_inet_epoch']}")
print(f"Łączny czas: {summ['total_time_s']/3600:.1f}h")
```

## 1. CIFAR-10 @ 32×32 — Accuracy

```python
h = cifar['history']
ep_c  = [x['epoch']     for x in h]
ta_c  = [x['train_acc'] for x in h]
te_c  = [x['test_acc']  for x in h]
lo_c  = [x['loss']      for x in h]

best_c_idx = int(np.argmax(te_c))

fig, ax = plt.subplots(figsize=(12, 5))
ax.plot(ep_c, ta_c, 'o-', label='Train Accuracy', color='#2E86AB', lw=2, markersize=5)
ax.plot(ep_c, te_c, 's-', label='Test Accuracy',  color='#A23B72', lw=2, markersize=5)
ax.scatter([ep_c[best_c_idx]], [te_c[best_c_idx]], s=250, color='gold', edgecolor='black',
           zorder=5, label=f'Best: {te_c[best_c_idx]:.2f}% (ep {ep_c[best_c_idx]})')
ax.set_xlabel('Epoch'); ax.set_ylabel('Accuracy (%)')
ax.set_title('ViT_Best @ 32×32 — 150 Epochs — CIFAR-10 — Accuracy',
             fontsize=12, fontweight='bold')
ax.legend(fontsize=10, loc='lower right'); ax.grid(alpha=0.3)
plt.tight_layout()
plt.savefig(f'{BASE_PATH}/150ep_cifar_accuracy.png', dpi=150, bbox_inches='tight')
plt.show()
```

## 2. CIFAR-10 @ 32×32 — Loss

```python
fig, ax = plt.subplots(figsize=(12, 5))
ax.plot(ep_c, lo_c, '^-', color='#F18F01', lw=2, markersize=5, label='Training Loss')
ax.axvline(ep_c[best_c_idx], color='#A23B72', linestyle='--', lw=1.5,
           label=f'Best test acc @ ep {ep_c[best_c_idx]}')
ax.set_xlabel('Epoch'); ax.set_ylabel('Loss')
ax.set_title('ViT_Best @ 32×32 — 150 Epochs — CIFAR-10 — Loss',
             fontsize=12, fontweight='bold')
ax.legend(fontsize=10); ax.grid(alpha=0.3)
plt.tight_layout()
plt.savefig(f'{BASE_PATH}/150ep_cifar_loss.png', dpi=150, bbox_inches='tight')
plt.show()
```

## 3. Imagenette2-320 @ 32×32 — Accuracy

```python
h2 = inet['history']
ep_i  = [x['epoch']     for x in h2]
ta_i  = [x['train_acc'] for x in h2]
te_i  = [x['test_acc']  for x in h2]
lo_i  = [x['loss']      for x in h2]

best_i_idx = int(np.argmax(te_i))

fig, ax = plt.subplots(figsize=(12, 5))
ax.plot(ep_i, ta_i, 'o-', label='Train Accuracy', color='#2E86AB', lw=2, markersize=5)
ax.plot(ep_i, te_i, 's-', label='Test Accuracy',  color='#A23B72', lw=2, markersize=5)
ax.scatter([ep_i[best_i_idx]], [te_i[best_i_idx]], s=250, color='gold', edgecolor='black',
           zorder=5, label=f'Best: {te_i[best_i_idx]:.2f}% (ep {ep_i[best_i_idx]})')
ax.set_xlabel('Epoch'); ax.set_ylabel('Accuracy (%)')
ax.set_title('ViT_Best @ 32×32 — 150 Epochs — Imagenette2-320 — Accuracy',
             fontsize=12, fontweight='bold')
ax.legend(fontsize=10, loc='lower right'); ax.grid(alpha=0.3)
plt.tight_layout()
plt.savefig(f'{BASE_PATH}/150ep_imagenet_accuracy.png', dpi=150, bbox_inches='tight')
plt.show()
```

## 4. Imagenette2-320 @ 32×32 — Loss

```python
fig, ax = plt.subplots(figsize=(12, 5))
ax.plot(ep_i, lo_i, '^-', color='#F18F01', lw=2, markersize=5, label='Training Loss')
ax.axvline(ep_i[best_i_idx], color='#A23B72', linestyle='--', lw=1.5,
           label=f'Best test acc @ ep {ep_i[best_i_idx]}')
ax.set_xlabel('Epoch'); ax.set_ylabel('Loss')
ax.set_title('ViT_Best @ 32×32 — 150 Epochs — Imagenette2-320 — Loss',
             fontsize=12, fontweight='bold')
ax.legend(fontsize=10); ax.grid(alpha=0.3)
plt.tight_layout()
plt.savefig(f'{BASE_PATH}/150ep_imagenet_loss.png', dpi=150, bbox_inches='tight')
plt.show()
```

## 5. Porównanie CIFAR-10 vs Imagenette — 150 epok

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 5))
fig.suptitle('ViT_Best @ 32×32 — 150 Epochs: CIFAR-10 vs Imagenette2-320',
             fontsize=13, fontweight='bold')

# Accuracy obu datasetów na jednym wykresie
axes[0].plot(ep_c, te_c, 's-', label='CIFAR-10',      color='#2E86AB', lw=2, markersize=4)
axes[0].plot(ep_i, te_i, 'o-', label='Imagenette',    color='#A23B72', lw=2, markersize=4)
axes[0].scatter([ep_c[best_c_idx]], [te_c[best_c_idx]], s=200, color='#2E86AB',
                edgecolor='black', zorder=5)
axes[0].scatter([ep_i[best_i_idx]], [te_i[best_i_idx]], s=200, color='#A23B72',
                edgecolor='black', zorder=5)
axes[0].set_xlabel('Epoch'); axes[0].set_ylabel('Test Accuracy (%)')
axes[0].set_title('Test Accuracy', fontsize=11, fontweight='bold')
axes[0].legend(fontsize=10); axes[0].grid(alpha=0.3)

# Loss obu datasetów
axes[1].plot(ep_c, lo_c, '^-', label='CIFAR-10',   color='#2E86AB', lw=2, markersize=4)
axes[1].plot(ep_i, lo_i, 'v-', label='Imagenette', color='#A23B72', lw=2, markersize=4)
axes[1].set_xlabel('Epoch'); axes[1].set_ylabel('Loss')
axes[1].set_title('Loss', fontsize=11, fontweight='bold')
axes[1].legend(fontsize=10); axes[1].grid(alpha=0.3)

plt.tight_layout()
plt.savefig(f'{BASE_PATH}/150ep_cifar_vs_imagenet.png', dpi=150, bbox_inches='tight')
plt.show()
```

## 6. Podsumowanie

```python
print('='*65)
print('PODSUMOWANIE — ViT_Best (patch_size=2) @ 32×32 — 150 EPOK')
print('='*65)

for name, d, ep, te in [('CIFAR-10', cifar, ep_c, te_c),
                         ('Imagenette2-320', inet, ep_i, te_i)]:
    best_idx = int(np.argmax(te))
    print(f'\n{name}:')
    print(f'  Best test acc  : {d["best_test_acc"]:.2f}%  @ epoch {d["best_epoch"]}')
    print(f'  Final test acc : {d["test_acc"]:.2f}%')
    print(f'  Final train acc: {d["train_acc"]:.2f}%')
    print(f'  Overfitting    : {d["train_acc"] - d["test_acc"]:.2f}%')
    print(f'  Czas treningu  : {d["train_time_s"]/3600:.2f}h  ({d["time_per_epoch"]:.1f}s/epoka)')

print(f'\nŁączny czas: {summ["total_time_s"]/3600:.1f}h')
print('='*65)
```
