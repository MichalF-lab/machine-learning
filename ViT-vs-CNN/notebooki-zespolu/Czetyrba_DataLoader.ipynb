```python
import torch
import torchvision
import torchvision.transforms as transforms
from torch.utils.data import DataLoader

# Obsługiwane datasety
DATASETS = {
    'cifar10':  {'classes': 10,  'img_size': 32},
    'cifar100': {'classes': 100, 'img_size': 32},
    'stl10':    {'classes': 10,  'img_size': 96},
}
```

```python
def data_loader(dataset='cifar10', batch_size=64, augment=False, path='./data'):
    """
    Ładuje wybrany dataset i zwraca (train_loader, test_loader, info).

    Args:
        dataset  : 'cifar10' | 'cifar100' | 'stl10'
        batch_size: rozmiar batcha
        augment  : czy stosować augmentację danych
        path     : katalog do zapisu danych

    Returns:
        train_loader, test_loader, info (dict z metadanymi)
    """
    assert dataset in DATASETS, f"Nieznany dataset: {dataset}. Dostępne: {list(DATASETS.keys())}"

    # --- Transformacje ---
    normalize = transforms.Normalize((0.5, 0.5, 0.5), (0.5, 0.5, 0.5))

    if augment:
        train_transforms = transforms.Compose([
            transforms.RandomHorizontalFlip(),
            transforms.RandomCrop(DATASETS[dataset]['img_size'], padding=4),
            transforms.ToTensor(),
            normalize,
        ])
    else:
        train_transforms = transforms.Compose([transforms.ToTensor(), normalize])

    test_transforms = transforms.Compose([transforms.ToTensor(), normalize])

    # --- Ładowanie datasetu ---
    # TODO: dodać obsługę 'stl10' i 'cifar100'
    if dataset == 'cifar10':
        train_set = torchvision.datasets.CIFAR10(root=path, train=True,  download=True, transform=train_transforms)
        test_set  = torchvision.datasets.CIFAR10(root=path, train=False, download=True, transform=test_transforms)
    elif dataset == 'cifar100':
        # TODO: zaimplementować
        raise NotImplementedError("CIFAR-100 – do zaimplementowania!")
    elif dataset == 'stl10':
        # TODO: zaimplementować
        raise NotImplementedError("STL-10 – do zaimplementowania!")

    train_loader = DataLoader(train_set, batch_size=batch_size, shuffle=True,  num_workers=2)
    test_loader  = DataLoader(test_set,  batch_size=batch_size, shuffle=False, num_workers=2)

    info = {
        'dataset':    dataset,
        'num_classes': DATASETS[dataset]['classes'],
        'img_size':   DATASETS[dataset]['img_size'],
        'train_size': len(train_set),
        'test_size':  len(test_set),
    }

    print(f"Dataset: {dataset} | Klasy: {info['num_classes']} | "
          f"Train: {info['train_size']} | Test: {info['test_size']}")
    return train_loader, test_loader, info
```

## Test – CIFAR-10 (powinien działać)

```python
train_loader, test_loader, info = data_loader(dataset='cifar10', batch_size=64, augment=True)
print(info)
```
